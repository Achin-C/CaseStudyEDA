# _DATA CLEANING_

#### _Listing all columns with null values with row numbers_

```python
nullseries = dfApplication_Data.isnull().sum()
print(nullseries[nullseries > 0])
```

#### _Removing columns with Null valued count more than 40% of all records_

```python
#Listing all columns with more than 40% of null values
columns = dfApplication_Data.columns
percent_missing = dfApplication_Data.isnull().sum() * 100 / len(dfApplication_Data)
missing_value_df = pd.DataFrame({'column_name': columns,'percent_missing': percent_missing})
missing_drop = list(missing_value_df[missing_value_df.percent_missing>40].column_name)
missing_drop
```

```python
#Dropping more than 40% null valued Columns and Naming it as dfApplication_Data_Standardized
dfApplication_Data_Standardized = dfApplication_Data.drop(missing_drop, axis=1)
```

```python
nullseries = dfApplication_Data_Standardized.isnull().sum() * 100 / len(dfApplication_Data_Standardized)
print(nullseries[(nullseries >0)])
```

```python
#describing dfApplication_Data_Standardized to see for means medians
print(dfApplication_Data_Standardized.describe())
```

```python
#filling na with mean or median
dfApplication_Data_Standardized.AMT_ANNUITY=dfApplication_Data_Standardized.AMT_ANNUITY.fillna(dfApplication_Data_Standardized.AMT_ANNUITY.mean())
dfApplication_Data_Standardized.AMT_GOODS_PRICE=dfApplication_Data_Standardized.AMT_GOODS_PRICE.fillna(dfApplication_Data_Standardized.AMT_GOODS_PRICE.mean())
dfApplication_Data_Standardized.CNT_FAM_MEMBERS=dfApplication_Data_Standardized.CNT_FAM_MEMBERS.fillna(dfApplication_Data_Standardized.CNT_FAM_MEMBERS.median())
dfApplication_Data_Standardized.EXT_SOURCE_2=dfApplication_Data_Standardized.EXT_SOURCE_2.fillna(dfApplication_Data_Standardized.EXT_SOURCE_2.median())
dfApplication_Data_Standardized.EXT_SOURCE_3=dfApplication_Data_Standardized.EXT_SOURCE_3.fillna(dfApplication_Data_Standardized.EXT_SOURCE_3.median())
dfApplication_Data_Standardized.OBS_30_CNT_SOCIAL_CIRCLE=dfApplication_Data_Standardized.OBS_30_CNT_SOCIAL_CIRCLE.fillna(dfApplication_Data_Standardized.OBS_30_CNT_SOCIAL_CIRCLE.median())
dfApplication_Data_Standardized.DEF_30_CNT_SOCIAL_CIRCLE=dfApplication_Data_Standardized.DEF_30_CNT_SOCIAL_CIRCLE.fillna(dfApplication_Data_Standardized.DEF_30_CNT_SOCIAL_CIRCLE.median())
dfApplication_Data_Standardized.OBS_60_CNT_SOCIAL_CIRCLE=dfApplication_Data_Standardized.OBS_60_CNT_SOCIAL_CIRCLE.fillna(dfApplication_Data_Standardized.OBS_60_CNT_SOCIAL_CIRCLE.median())
dfApplication_Data_Standardized.DEF_60_CNT_SOCIAL_CIRCLE =dfApplication_Data_Standardized.DEF_60_CNT_SOCIAL_CIRCLE.fillna(dfApplication_Data_Standardized.DEF_60_CNT_SOCIAL_CIRCLE.median())
dfApplication_Data_Standardized.DAYS_LAST_PHONE_CHANGE=dfApplication_Data_Standardized.DAYS_LAST_PHONE_CHANGE.fillna(dfApplication_Data_Standardized.DAYS_LAST_PHONE_CHANGE.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_HOUR=dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_HOUR.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_DAY =dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_DAY.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_DAY.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_WEEK=dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_WEEK.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_MON=dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_MON.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_MON.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_QRT=dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_QRT.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_QRT.median())
dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_YEAR=dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(dfApplication_Data_Standardized.AMT_REQ_CREDIT_BUREAU_YEAR.median())

#NAME_TYPE_SUITE filling with "Unknown" value for null values
dfApplication_Data_Standardized.NAME_TYPE_SUITE=dfApplication_Data_Standardized.NAME_TYPE_SUITE.fillna('Unknown')
dfApplication_Data_Standardized.OCCUPATION_TYPE=dfApplication_Data_Standardized.OCCUPATION_TYPE.fillna('Unknown')
```

### _1 . Analysis Based on Gender_

### 2.Analysis on AGE of Applicant derived from DAYS_BIRTH vs TARGET

```python
#Converting days_birth into age and then bucketing them
dfApplication_Data_Standardized['AGE']=dfApplication_Data_Standardized.DAYS_BIRTH.apply(lambda x: round(abs(x/365)),2)
# Binning the data based on age_groups
dfApplication_Data_Standardized['DAYS_BIRTH_IN_AGE_GROUPS']=pd.cut(dfApplication_Data_Standardized.AGE, bins=[0,  30, 40, 50, 60, 9999], labels=[ '0-30', '31-40', '41-50', '51-60', '61+'])
```

```python
dfApplication_Data_Standardized.DAYS_BIRTH_IN_AGE_GROUPS.value_counts(normalize=True).plot.bar()
plt.show()
```

```python
dfApplication_Data_Standardized.groupby('DAYS_BIRTH_IN_AGE_GROUPS')['TARGET'].mean().plot.bar()
plt.show()
```

```python
a=pd.pivot_table(data=dfApplication_Data_Standardized, columns='TARGET', index='DAYS_BIRTH_IN_AGE_GROUPS', values='TARGET_1', aggfunc='count').sort_values(1, ascending=False)
a.plot.bar(figsize=[8,6], width=.4)
plt.show()
```

#### <font color='green'>We can see from the above three graphs that the number of entries is lowest in the 61+ age group bit it also has the least percentage of defaulters.</font>

#### <font color='green'>No. of defaulters with respect to the total number of applications is decreasing with increase in age groups.</font>

### 8. ORGANIZATION_TYPE vs TARGET

```python
dfApplication_Data_Standardized.ORGANIZATION_TYPE.unique()
```

```python
def type_parser(i):
    if ':' in i.lower():
        return i.split(':')[0]
    elif 'type' in i.lower():
        return i.split()[0]
    else:
        return i
```

```python
dfApplication_Data_Standardized['ORGANIZATION_TYPE_Standardized']=dfApplication_Data_Standardized.ORGANIZATION_TYPE.apply(lambda x: type_parser(x))
```

```python
dfApplication_Data_Standardized.ORGANIZATION_TYPE_Standardized.value_counts().plot.barh(figsize=[20,8])
plt.show()
```

```python
a=pd.pivot_table(data=dfApplication_Data_Standardized, columns='TARGET', index='ORGANIZATION_TYPE_Standardized', values='TARGET_1', aggfunc='count').sort_values(1, ascending=False)
a.apply(lambda x: x/x.sum(), axis=1).sort_values(1).plot(kind='barh', stacked=True, figsize=[20,8])
plt.show()
```

<font color='green'>we can see that business, self-employed, industy,trade etc have most applications in descending order. Religion, cleaning and legal services have least no. of applications</font>

<font color='green'>Security ministries, university, police, military etc are least defaulting while resuarants, construction, cleaning works etc are more likely to default</font>
