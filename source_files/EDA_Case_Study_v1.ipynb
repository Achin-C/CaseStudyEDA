### AGE vs TARGET

```python
appl_df=pd.read_csv('application_data.csv', encoding='ISO-8859–1')
```

```python
# Converting days_birth into age and then bucketing them.
appl_df['AGE']=appl_df.DAYS_BIRTH.apply(lambda x: round(abs(x/365)),2)
appl_df.head(5)
```

```python
# Binning the data based on age_groups
appl_df['age_groups']=pd.cut(appl_df.AGE, bins=[0, 20, 30, 40, 50, 60, 9999], labels=['<20', '21-30', '31-40', '41-50', '51-60', '61+'])
```

```python
# Making desirable entries as 1(i.e Non defaulters) and defaulters as 0 for ease of analysis.
appl_df['TARGET']=appl_df.TARGET.apply(lambda x: 1 if x==0 else 0)
```

```python
appl_df.AGE.plot.box()
```

```python
# Looks like good data.
```

```python
appl_df.groupby('age_groups')['TARGET'].mean().plot.bar()
plt.show()
appl_df.age_groups.value_counts(normalize=True).plot.bar()
plt.show()
```

```python
appl_df['TARGET_1']=appl_df['TARGET']
```

```python
a=pd.pivot_table(data=appl_df, columns='TARGET', index='age_groups', values='TARGET_1', aggfunc='count').sort_values(1, ascending=False)
a.plot.bar()
plt.show()
a.plot.bar(stacked=True)
plt.show()
a.apply(lambda x: x/x.sum(), axis=1).sort_values(1).plot(kind='bar', stacked=True, figsize=[8,6], width=.4)
plt.show()
```

#### <font color='green'>We can see from the above three graphs that the number of entries is lowest in the 61+ age group bit it also has the least percentage of defaulters.</font>

#### <font color='green'>No. of defaulters with respect to the total number of applications is decreasing with increase in age groups.</font>

#### <font color='green'>Highest number of applications are from age-group 31-40 and it reduces as we move to either of the two sides.</font>

## ORGANIZATION_TYPE vs TARGET

```python
appl_df.ORGANIZATION_TYPE.unique()
```

```python
def type_parser(i):
    if ':' in i.lower():
        return i.split(':')[0]
    elif 'type' in i.lower():
        return i.split()[0]
    else:
        return i
```

```python
appl_df['ORGANIZATION_TYPE1']=appl_df.ORGANIZATION_TYPE.apply(lambda x: type_parser(x))
```

```python
appl_df.ORGANIZATION_TYPE1.value_counts().plot.barh(figsize=[20,8])
```

```python
# Getting % of xna entries.
appl_df[appl_df['ORGANIZATION_TYPE1']=='XNA'].ORGANIZATION_TYPE1.count()*100/appl_df.ORGANIZATION_TYPE1.count()
```

```python
# xna comprises of about 18% of the overall organisation type. Since we cannot fill in these much values with anything else, i am going to ignore them for now.
appl_df1=appl_df[~(appl_df.ORGANIZATION_TYPE1=='XNA')]
```

```python
appl_df1.ORGANIZATION_TYPE1.unique()
```

```python
appl_df1.ORGANIZATION_TYPE1.unique()
appl_df1.ORGANIZATION_TYPE1.value_counts().plot.pie(figsize=[20,8])
plt.show()
appl_df1.ORGANIZATION_TYPE1.value_counts().plot.barh(figsize=[20,8])
plt.show()
a=pd.pivot_table(data=appl_df1, columns='TARGET', index='ORGANIZATION_TYPE1', values='TARGET_1', aggfunc='count').sort_values(1, ascending=False)
a.plot.barh(figsize=[20,8])
plt.show()
a.plot.barh(stacked=True, figsize=[20,8])
plt.show()
a.apply(lambda x: x/x.sum(), axis=1).sort_values(1).plot(kind='barh', stacked=True, figsize=[20,8])
plt.show()
```

<font color='green'>we can see that business, self-employed, industy,trade etc have most applications in descending order. Religion, cleaning and legal services have least no. of applications</font>

<font color='green'>Security ministries, university, police, military etc are least defaulting while resuarants, construction, cleaning works etc are more likely to default</font>
